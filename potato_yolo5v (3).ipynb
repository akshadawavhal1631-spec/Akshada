```python
from google.colab import files
from ultralytics import YOLO

# Load the trained model
# Replace 'path/to/your/trained_model.pt' with the actual path to your best.pt or last.pt file
# You can find this in the runs/detect/train/weights directory from your training run.
# For example, if your training run was in 'runs/detect/train', the model would be in 'runs/detect/train/weights/best.pt'
model = YOLO('/content/runs/detect/train/weights/best.pt') # Assuming the best model is saved here
```

```python
uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

# Run inference on the uploaded image
results = model('/content/' + list(uploaded.keys())[0])
```

```python
from IPython.display import Image, display

# Display the annotated image directly from the results object
display(results[0].plot())

# Print the predicted classes and confidence scores
for r in results:
    for c in r.boxes.cls:
        print(f"Predicted class: {model.names[int(c)]}")
    for s in r.boxes.conf:
        print(f"Confidence score: {float(s):.2f}")
```
